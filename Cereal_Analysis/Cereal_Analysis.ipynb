# Analysis of Popular Cereals

## Introduction:
The following is an analysis of several traits in both hot and cold cereals that are popular in the United States. These traits include nutrition facts, rating, temperature, etc. and how they may relate


## Macros:
This plot contains the distributions of protein, fat, and carbohydrates contained in the cereals in the dataframe.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np


df = pd.read_csv('cereal.csv')

fig, ax = plt.subplots()
fig = sns.kdeplot(df['protein'], ax=ax, color="r", label='protein')
fig = sns.kdeplot(df['fat'], ax=ax,color="b", label = 'fat')
fig = sns.kdeplot(df['carbo'], ax=ax,color="g", label = 'carbs')
ax.set_xlabel('macros')
ax.legend()
plt.show()
```

## Sugar Content:

```python
# Cereals with the most sugars
sugarDf = df.loc[:,['name', 'sugars']]
sugarDf.sort_values('sugars', ascending = False).head(10)
```

```python
# Cereals with the least sugars
sugarDf.sort_values('sugars', ascending = True).head(10)
```

## Relationships Between Nutrients and Rating:
How are the contents of different nutrients and the ratings of cereals related in the United States?

```python
# Relationship between cereal carbohydrates and rating
sns.regplot(data=df, x = 'carbo', y = 'rating')
plt.show()
x = pd.Series(df['carbo'])
y = pd.Series(df['rating'])
rVal = x.corr(y)
print(rVal)
```

```python
# Relationship between cereal sugar content and rating
sns.regplot(data=df, x = 'sugars', y = 'rating')
plt.show()
x = pd.Series(df['sugars'])
y = pd.Series(df['rating'])
rVal = x.corr(y)
print(rVal)
```

```python
# Relationship between cereal calories and rating
sns.regplot(data=df, x = 'calories', y = 'rating')
plt.show()
x = pd.Series(df['calories'])
y = pd.Series(df['rating'])
rVal = x.corr(y)
print(rVal)
```

```python
# Relationship between cereal protein and rating
sns.regplot(data=df, x = 'protein', y = 'rating')
plt.show()
x = pd.Series(df['protein'])
y = pd.Series(df['rating'])
rVal = x.corr(y)
print(rVal)
```

```python
# Relationship between cereal fat and rating
sns.regplot(data=df, x = 'fat', y = 'rating')
plt.show()
x = pd.Series(df['fat'])
y = pd.Series(df['rating'])
rVal = x.corr(y)
print(rVal)
```

By calculating the 'r' value for each graph, I found that the amount of sugar in a cereal has the strongest relationship with its rating.

## Average Rating for Hot and Cold Cereals
This cell calculates the average rating of hot cereals and cold cereals in the dataframe.

```python
df.loc[:,['type', 'rating']]
hotArray = []
coldArray = []
for i in range(len(df.loc[:,'type'])):
  if(df.loc[i,'type'] == 'C'):
    coldArray.append(df.loc[i,'rating'])
  elif(df.loc[i,'type'] == 'H'):
    hotArray.append(df.loc[i,'rating'])

hotArray = np.array(hotArray)
coldArray = np.array(coldArray)
hotAvg = np.mean(hotArray)
coldAvg = np.mean(coldArray)

print('Hot average rating:', hotAvg)
print('Cold average rating:', coldAvg)
```
